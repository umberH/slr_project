```python
import pandas as pd # for data manipulation
import numpy as np # for numerical computation
import os 
from sklearn.preprocessing import LabelEncoder

```

# Data Ingestion


```python
#print(os.listdir())
data = pd.read_csv("data/train.csv")
data
```

```python

le = LabelEncoder()

for cat_var in ['Gender', 'Vehicle_Damage']:
    data[cat_var] = le.fit_transform(data[cat_var])  
```

```python
data
```

```python
X = ["Gender", "Age"]

mu = data.groupby("Previously_Insured")[X].mean()
var = data.groupby("Previously_Insured")[X].var()

norm_diff = ((mu - mu.loc[1])/
                np.sqrt((var + var.loc[1])/2))

norm_diff
```
